```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
#its a nominal value so values should be expanded
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int) # expanded columns are boolean so converted into int
dataset.columns
```

```python
#split input and output
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
 from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
'max_features': ['sqrt', 'log2', None],
'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(independent, dependent)
```

```python
re=grid.cv_results_
#print(re)
grid_predictions = grid.predict(independent) # doubt independent will have all the data or only folded data
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(dependent,grid_predictions)
print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)
```
